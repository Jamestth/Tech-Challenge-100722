```python
import requests
import pandas as pd
```

```python
endpoint = "https://api.covid19api.com/country/singapore/status/confirmed"

# Time range not a necessary parameter. Without indicating, all data is extracted. 
# #"2020-03-01T00:00:00Z"
# data_postfix = "T00:00:00Z"
# start_date = "2020-03-01"
# end_date = "2022-07-9"
# params = {"from": start_date + data_postfix, "to": end_date + data_postfix }
req = requests.get(endpoint)#, params=params)
```

```python
# Convert to df
df = pd.DataFrame(req.json())
df
```

```python
# Save API_data
df.to_csv("API_data.csv", index = False)
```
